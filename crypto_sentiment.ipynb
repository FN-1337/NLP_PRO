## 1. Sentiment Analysis

Use the [newsapi](https://newsapi.org/) to pull the latest news articles for Bitcoin and Ethereum and create a DataFrame of sentiment scores for each coin.

Use descriptive statistics to answer the following questions:
1. Which coin had the highest mean positive score?
2. Which coin had the highest negative score?
3. Which coin had the highest positive score?

```python
# Initial imports
import os
import pandas as pd
from dotenv import load_dotenv
from newsapi import NewsApiClient
import nltk as nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

%matplotlib inline
```

```python
# Read your api key environment variable
load_dotenv()
news_api_key = os.getenv("NEWS_API_KEY")
```

```python
# Create a newsapi client
newsapi = NewsApiClient(api_key=news_api_key)
```

```python
# Fetch the Bitcoin news articles
bitcoin_articles = newsapi.get_everything(q='bitcoin',
                                      language='en',
                                      sort_by='relevancy')
print(f"Total articles about bitcoin: {bitcoin_articles['totalResults']}")
bitcoin_articles["articles"][0]
```

```python
# Fetch the Ethereum news articles
ethereum_articles = newsapi.get_everything(q='Ethereum',
                                      language='en',
                                      sort_by='relevancy')
print(f"Total articles about ethereum: {ethereum_articles['totalResults']}")
ethereum_articles["articles"][0]
```

```python
# Create the Bitcoin sentiment scores DataFrame
bitcoin_sentiments = []

for article in bitcoin_articles["articles"]:
    try:
        text = article["content"]
        sentiment = analyzer.polarity_scores(text)
        compound = sentiment["compound"]
        pos = sentiment["pos"]
        neu = sentiment["neu"]
        neg = sentiment["neg"]
        
        bitcoin_sentiments.append({
            "text": text,
            "compound": compound,
            "positive": pos,
            "negative": neg,
            "neutral": neu
            
        })
        
    except AttributeError:
        pass

bitcoin_df = pd.DataFrame(bitcoin_sentiments)
cols = [ "text", "compound","positive", "negative", "neutral"]
bitcoin_df = bitcoin_df[cols]
bitcoin_df.head()
```

```python
# Create the Ethereum sentiment scores DataFrame
ethereum_sentiments = []

for article in ethereum_articles["articles"]:
    try:
        text = article["content"]
        sentiment = analyzer.polarity_scores(text)
        compound = sentiment["compound"]
        pos = sentiment["pos"]
        neu = sentiment["neu"]
        neg = sentiment["neg"]
        
        ethereum_sentiments.append({
            "text": text,
            "compound": compound,
            "positive": pos,
            "negative": neg,
            "neutral": neu
            
        })
        
    except AttributeError:
        pass

ethereum_df = pd.DataFrame(ethereum_sentiments)
cols = [
    "text", "compound", "positive", "negative", "neutral"]
ethereum_df = ethereum_df[cols]
ethereum_df.head()
```

```python
# Describe the Bitcoin Sentiment
bitcoin_sent=bitcoin_df.describe()
bitcoin_sent
```

```python
# Describe the Ethereum Sentiment
ethereum_sent=ethereum_df.describe()
ethereum_sent
```

###   Tokenizer

In this section, you will use NLTK and Python to tokenize the text for each coin. Be sure to:
1. Lowercase each word.
2. Remove Punctuation.
3. Remove Stopwords.

```python
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from string import punctuation
import re
```

```python
# Instantiate the lemmatizer
lemmatizer = WordNetLemmatizer()

# Create a list of stopwords
sw = set(stopwords.words('english'))

# Expand the default stopwords list if necessary
sw_addon = {'said', 'sent', 'found', 'including', 'today', 'announced', 'week', 'basically', 'also', 'weekly', 'bitcoin', 'ethereum', 'following', 'char', 'hi','ha','wa','tha'}
```

```python
# Complete the tokenizer function
def tokenizer(text):
    sw= set(stopwords.words('english'))
    regex= re.compile("[^a-zA-Z ]")
    
    re_clean = regex.sub('', str(text))
    words= word_tokenize(re_clean)
    lem=[lemmatizer.lemmatize(word) for word in words]
    tokens= [word.lower() for word in lem if word.lower() not in sw.union(sw_addon) ]
    
    return tokens
```

```python
# Create a new tokens column for Bitcoin
bitcoin_df["tokens"] = bitcoin_df.text.apply(tokenizer)
bitcoin_df.head()
```

```python
# Create a new tokens column for Ethereum
ethereum_df["tokens"] = ethereum_df.text.apply(tokenizer)
ethereum_df.head()
```

### NGrams and Frequency Analysis

In this section you will look at the ngrams and word frequency for each coin. 

1. Use NLTK to produce the n-grams for N = 2. 
2. List the top 10 words for each coin. 

```python
from collections import Counter
from nltk import ngrams
```

```python
# Generate the Bitcoin N-grams where N=2
bitcoin_ngrams = [ Counter(ngrams(token, n=2)) for token in bitcoin_df["tokens"].values ]
bitcoin_ngrams[0]
```

```python
# Generate the Ethereum N-grams where N=2
ethereum_ngrams = [ Counter(ngrams(token, n=2)) for token in ethereum_df["tokens"].values ]
ethereum_ngrams[0]
```

```python
# Function token_count generates the top 10 words for a given coin
def token_count(tokens, N=3):
    """Returns the top N tokens from the frequency count"""
    return Counter(tokens).most_common(N)
```

```python
# Use token_count to get the top 10 words for Bitcoin
all_words_bitcoin = []
for token in bitcoin_df.tokens:
    all_words_bitcoin += token

token_count(all_words_bitcoin, 10)
```

```python
# Use token_count to get the top 10 words for Ethereum
all_words_ethereum = []
for token in ethereum_df.tokens:
    all_words_ethereum += token

token_count(all_words_ethereum, 10)
```

## 3. Named Entity Recognition

In this section, you will build a named entity recognition model for both Bitcoin and Ethereum, then visualize the tags using SpaCy.

```python
import spacy
from spacy import displacy
```

```python
# Download the language model for SpaCy
#!python -m spacy download en_core_web_sm
```

```python
# Load the spaCy model
nlp = spacy.load('en_core_web_sm')
```

---
### Bitcoin NER

```python
# Concatenate all of the Bitcoin text together
bitcoin_corpus = ' '.join(bitcoin_df["text"].values)
```

```python
# Run the NER processor on all of the text
btc_ner = nlp(bitcoin_corpus)

# Add a title to the document
btc_ner.user_data["title"] = "All the Article Text on Bitcoin"
```

```python
# Render the visualization
displacy.render(btc_ner, style='ent')
```

```python
# List all Entities
bitcoin_entities_list = [ent.text for ent in btc_ner.ents]
bitcoin_entities_set = set(bitcoin_entities_list)
bitcoin_entities_list
```

```python
# Render the visualization
displacy.render(eth_ner, style='ent')
```

```python
# List all Entities
ethereum_entities_list = [ent.text for ent in eth_ner.ents]
ethereum_entities_set = set(ethereum_entities_list)
ethereum_entities_list
```
